```python
import pandas as pd
import os
```

```python
FOLDER_PATH = "Dataset_test"
FILE_NAME = "without_price.csv"
file_path = os.path.join(FOLDER_PATH, FILE_NAME)
df = pd.read_csv(file_path, dtype={"Timestamp": "int64"})
```

We want to use the language model called Finalcial BERT. This one, compared to the textblob one doesnt recommend removing stopwords and punctuation. It also recommends to not remove uppercase letters, so all the cleaning we did for textblob isn't necessary.

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch
```

```python
# Load the tokenizer and model for Financial BERT
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
model = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone')

def get_sentiment_score(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
    outputs = model(**inputs)
    logits = outputs.logits
    probs = torch.nn.functional.softmax(logits, dim=-1)
    
    positive_score = probs[0][0].item()
    # Scale the score to be between 0 and 1
    sentiment_score = positive_score
    
    return sentiment_score

def analyze_sentiment(row):
    text = row['Text']
    sentiment_score = get_sentiment_score(text)
    return sentiment_score

df['Sentiment Score'] = df.apply(analyze_sentiment, axis=1)
```

```python
follower_counts = {
    'AlanSantana': 132467,
    'MoralDisciple': 1127,
    'RocketBomb': 11991,
    'CobraVanguard': 15073,
    'CryptoColugo' : 4724,
    'CRYPTOMOJO_TA': 11944,
    'FieryTrading' : 49050,
    'MoralDisciple' : 1127,
    'weslad' : 23976,
    'without_worries' : 12485,
    'Xanrox' : 10455
}

def get_follower_count(trader):
    return follower_counts.get(trader, 0)

df['Follower Count'] = df['Trader'].apply(get_follower_count)
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# Function to get trader colors
def get_trader_colors(traders):
    unique_traders = list(set(traders))
    colors = plt.get_cmap('tab10', len(unique_traders))  # Get a colormap with enough colors
    trader_color_map = {trader: colors(i) for i, trader in enumerate(unique_traders)}
    return trader_color_map

# Function to create the BTC graph
def create_btc_graph(topic_df, follower_counts):
    G = nx.DiGraph()
    
    topic_node = 'BTC'
    G.add_node(topic_node, label=topic_node, layer=0, color='gray', weight=0.5, sentiment_score=0.5)
    
    topic_df = topic_df.sort_values(by='Timestamp')
    last_post_label = {}
    
    trader_colors = get_trader_colors(topic_df['Trader'])
    
    for _, row in topic_df.iterrows():
        post_label = f"Post_{row['Timestamp']}"
        title = row['Title']
        trader = row['Trader']
        post_type = row['Post type']
        sentiment_score = row['Sentiment Score']
        
        # Determine node color based on sentiment score
        if sentiment_score > 0.5:
            node_color = 'green'
        elif sentiment_score < 0.5:
            node_color = 'red'
        else:
            node_color = 'gray'
        
        # Initial posts connect to the topic node
        if post_type == 'Initial post':
            G.add_node(post_label, label=trader, layer=1, color=node_color, weight=follower_counts[trader], sentiment_score=sentiment_score)
            G.add_edge(post_label, topic_node, color='black')
            last_post_label[title] = post_label
        else:
            if title in last_post_label:
                previous_post_label = last_post_label[title]
                G.add_node(post_label, label=trader, layer=G.nodes[previous_post_label]['layer'] + 1, color=node_color, weight=follower_counts[trader], sentiment_score=sentiment_score)
                
                previous_color = G.nodes[previous_post_label]['color']
                edge_color = 'green' if previous_color == node_color else 'red'
                
                G.add_edge(post_label, previous_post_label, color=edge_color)
                last_post_label[title] = post_label

    # Calculate the average sentiment score for BTC
    sentiment_scores = [data['sentiment_score'] for node, data in G.nodes(data=True) if node != 'BTC']
    if sentiment_scores:
        average_sentiment = sum(sentiment_scores) / len(sentiment_scores)
        G.nodes[topic_node]['sentiment_score'] = average_sentiment
        if average_sentiment > 0.5:
            G.nodes[topic_node]['color'] = 'green'
        elif average_sentiment < 0.5:
            G.nodes[topic_node]['color'] = 'red'
        else:
            G.nodes[topic_node]['color'] = 'gray'
        
        # Update edges connected to BTC
        for child in G.predecessors(topic_node):  # Use predecessors instead of successors
            child_color = G.nodes[child]['color']
            edge_color = 'green' if G.nodes[topic_node]['color'] == child_color else 'red'
            G.edges[(child, topic_node)]['color'] = edge_color
    
    pos = nx.multipartite_layout(G, subset_key="layer", align="vertical")

    # Adjust positions to create more space between nodes
    for k, v in pos.items():
        pos[k] = (v[0] * 1.5, v[1] * 2)

    plt.figure(figsize=(16, 12))
    
    node_colors = [G.nodes[node]['color'] for node in G.nodes]
    edge_colors = [G[u][v]['color'] for u, v in G.edges]
    
    nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'label'), node_size=3000, node_color=node_colors, font_size=10, font_color="black", font_weight="bold", edge_color=edge_colors, arrows=True)
    plt.title('BTC Topic Graph')
    plt.show()


# Replace btc_df with the appropriate DataFrame variable name
create_btc_graph(btc_df[:13], follower_counts)
```

```python
def get_trader_colors(traders):
    unique_traders = list(set(traders))
    colors = plt.get_cmap('tab10', len(unique_traders))  # Get a colormap with enough colors
    trader_color_map = {trader: colors(i) for i, trader in enumerate(unique_traders)}
    return trader_color_map

def create_btc_graph_and_bag(topic_df, follower_counts, bag_filename):
    # Scale follower counts to [0, 1]
    followers = np.array(list(follower_counts.values())).reshape(-1, 1)
    scaler = MinMaxScaler()
    scaled_followers = scaler.fit_transform(followers)
    scaled_follower_counts = {trader: scaled_followers[i][0] for i, trader in enumerate(follower_counts)}
    
    G = nx.DiGraph()
    
    topic_node = 'BTC'
    G.add_node(topic_node, label=topic_node, layer=0, color='gray', weight=0.5, sentiment_score=0.5)
    
    topic_df = topic_df.sort_values(by='Timestamp')
    last_post_label = {}
    
    trader_colors = get_trader_colors(topic_df['Trader'])
    
    for _, row in topic_df.iterrows():
        post_label = f"Post_{row['Timestamp']}"
        title = row['Title']
        trader = row['Trader']
        post_type = row['Post type']
        sentiment_score = row['Sentiment Score']
        
        # Determine node color based on sentiment score
        if sentiment_score > 0.5:
            node_color = 'green'
        elif sentiment_score < 0.5:
            node_color = 'red'
        else:
            node_color = 'gray'
        
        # Initial posts connect to the topic node
        if post_type == 'Initial post':
            weight = (sentiment_score + scaled_follower_counts[trader]) / 2  # Combine sentiment score and scaled follower count
            G.add_node(post_label, label=trader, layer=1, color=node_color, weight=weight, sentiment_score=sentiment_score)
            G.add_edge(post_label, topic_node, color='black')
            last_post_label[title] = post_label
        else:
            if title in last_post_label:
                previous_post_label = last_post_label[title]
                weight = (sentiment_score + scaled_follower_counts[trader]) / 2  # Combine sentiment score and scaled follower count
                G.add_node(post_label, label=trader, layer=G.nodes[previous_post_label]['layer'] + 1, color=node_color, weight=weight, sentiment_score=sentiment_score)
                
                previous_color = G.nodes[previous_post_label]['color']
                edge_color = 'green' if previous_color == node_color else 'red'
                
                G.add_edge(post_label, previous_post_label, color=edge_color)
                last_post_label[title] = post_label

    # Update edges connected to BTC
    for child in G.predecessors(topic_node):
        child_color = G.nodes[child]['color']
        edge_color = 'green' if G.nodes[topic_node]['color'] == child_color else 'red'
        G.edges[(child, topic_node)]['color'] = edge_color
    
    pos = nx.multipartite_layout(G, subset_key="layer", align="vertical")

    # Adjust positions to create more space between nodes
    for k, v in pos.items():
        pos[k] = (v[0] * 1.5, v[1] * 2)

    plt.figure(figsize=(16, 12))
    
    node_colors = [G.nodes[node]['color'] for node in G.nodes]
    edge_colors = [G[u][v]['color'] for u, v in G.edges]
    
    nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'label'), node_size=3000, node_color=node_colors, font_size=10, font_color="black", font_weight="bold", edge_color=edge_colors, arrows=True)
    plt.title('BTC Topic Graph')
    plt.show()
    
    # Create .bag file
    with open(bag_filename, 'w') as f:
        for node, data in G.nodes(data=True):
            f.write(f"arg({node}, {data['weight']}).\n")
        for u, v, data in G.edges(data=True):
            if data['color'] == 'green':
                f.write(f"sup({u}, {v}).\n")
            else:
                f.write(f"att({u}, {v}).\n")

# Replace btc_df with the appropriate DataFrame variable name
create_btc_graph_and_bag(btc_df, follower_counts, 'BTC_graph.bag')
```
