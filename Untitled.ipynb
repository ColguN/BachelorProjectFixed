```python
import pandas as pd
import numpy as np
import os
```

```python
# Define the file path
FOLDER_PATH = "Dataset_test"
file_path = os.path.join(FOLDER_PATH, "dataset-final.csv")

# Load the CSV file into a DataFrame
df_final = pd.read_csv(file_path)
```

```python
from textblob import TextBlob as tb
# Polarity means the sentiment of the text (in range of -1 to 1)
df_final["Polarity"] = df_final["Lemmatized"].apply(lambda x: tb(x).sentiment[0])
df_final["Polarity"]
```

```python
# subjectivity means the degree to which the text is subjective (in range of 0 to 1)
df_final["Subjectivity"] = df_final["Lemmatized"].apply(lambda x: tb(x).sentiment[1])
df_final["Subjectivity"]
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Add a column to determine buy/sell decision based on polarity
df_final['Decision'] = df_final['Polarity'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# Filter for the topic "Solana"
solana_df = df_final[df_final['Topic'] == 'SOL']
```

```python
if not solana_df.empty:
    # Function to create and visualize graph for Solana topic
    def create_solana_graph(topic_df):
        G = nx.DiGraph()
        
        # Add the decision node
        decision_node = 'Decision'
        G.add_node(decision_node, label=decision_node, layer=0)
        
        # Add the topic node
        topic_node = 'Solana'
        G.add_node(topic_node, label=topic_node, layer=1)
        G.add_edge(decision_node, topic_node)
        
        # Sort posts by timestamp
        topic_df = topic_df.sort_values(by='Timestamp')
        
        # Dictionary to keep track of the last post for each title
        last_post_label = {}
        
        for _, row in topic_df.iterrows():
            post_label = f"Post_{row['Timestamp']}"
            title = row['Title']
            post_type = row['Post type']
            
            if post_type == 'Initial post':
                # Initial post node
                G.add_node(post_label, label=title, layer=2)
                G.add_edge(topic_node, post_label)
                last_post_label[title] = post_label
            else:
                # Update post node
                G.add_node(post_label, label=title, layer=3)
                if title in last_post_label:
                    initial_post_label = last_post_label[title]
                    G.add_edge(initial_post_label, post_label)
                else:
                    G.add_edge(topic_node, post_label)
        
        # Define the positions of nodes for visualization
        pos = nx.multipartite_layout(G, subset_key="layer")
        
        # Draw the graph
        plt.figure(figsize=(12, 8))
        nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'label'), node_size=3000, node_color="skyblue", font_size=10, font_color="black", font_weight="bold")
        plt.title('Decision and Post Relationship Graph for Solana')
        plt.show()
    
    # Create and visualize graph for Solana topic
    create_solana_graph(solana_df)
else:
    print("No posts for the topic 'Solana'.")
```

```python
if not solana_df.empty:
    # Define a function to generate a color map for traders
    def get_trader_colors(traders):
        unique_traders = list(set(traders))
        colors = plt.get_cmap('tab10', len(unique_traders))  # Get a colormap with enough colors
        trader_color_map = {trader: colors(i) for i, trader in enumerate(unique_traders)}
        return trader_color_map
    
    # Function to create and visualize graph for Solana topic
    def create_solana_graph(topic_df):
        G = nx.DiGraph()
        
        # Add the decision node
        decision_node = 'Decision'
        G.add_node(decision_node, label=decision_node, layer=0, color='gray', decision=None)
        
        # Add the topic node
        topic_node = 'Solana'
        G.add_node(topic_node, label=topic_node, layer=1, color='gray', decision=None)
        G.add_edge(topic_node, decision_node, color='black')
        
        # Sort posts by timestamp
        topic_df = topic_df.sort_values(by='Timestamp')
        
        # Dictionary to keep track of the last post for each title
        last_post_label = {}
        layer_counter = 2
        
        # Get the color map for traders
        trader_colors = get_trader_colors(topic_df['Trader'])
        
        for _, row in topic_df.iterrows():
            post_label = f"Post_{row['Timestamp']}"
            title = row['Title']
            trader = row['Trader']
            post_type = row['Post type']
            decision = row['Decision']
            
            if post_type == 'Initial post':
                # Initial post node
                G.add_node(post_label, label=trader, layer=layer_counter, color=trader_colors[trader], decision=decision)
                G.add_edge(post_label, topic_node, color='black')
                last_post_label[title] = post_label
            else:
                # Update post node
                if title in last_post_label:
                    previous_post_label = last_post_label[title]
                    layer_counter += 1
                    G.add_node(post_label, label=trader, layer=layer_counter, color=trader_colors[trader], decision=decision)
                    
                    # Determine edge color based on decision
                    previous_decision = G.nodes[previous_post_label]['decision']
                    edge_color = 'green' if previous_decision == decision else 'red'
                    
                    G.add_edge(post_label, previous_post_label, color=edge_color)
                    last_post_label[title] = post_label
        
        # Define the positions of nodes for visualization
        pos = nx.multipartite_layout(G, subset_key="layer", align="vertical")
        
        # Draw the graph
        plt.figure(figsize=(16, 12))
        
        # Extract node colors
        node_colors = [nx.get_node_attributes(G, 'color').get(node, 'gray') for node in G.nodes]
        edge_colors = [G[u][v]['color'] for u, v in G.edges]
        
        nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'label'), node_size=3000, node_color=node_colors, font_size=10, font_color="black", font_weight="bold", edge_color=edge_colors, arrows=True)
        plt.title('Decision and Post Relationship Graph for Solana')
        plt.show()
    
    # Create and visualize graph for Solana topic
    create_solana_graph(solana_df)
else:
    print("No posts for the topic 'Solana'.")
```
