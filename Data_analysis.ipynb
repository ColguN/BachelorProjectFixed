We want to use the language model called Finalcial BERT. This one, compared to the textblob one doesnt recommend removing stopwords and punctuation. It also recommends to not remove uppercase letters, so all the cleaning we did for textblob isn't necessary.

```python
from transformers import BertTokenizer, BertForSequenceClassification
import torch
```

```python
# Load the tokenizer and model for Financial BERT
tokenizer = BertTokenizer.from_pretrained('yiyanghkust/finbert-tone')
model = BertForSequenceClassification.from_pretrained('yiyanghkust/finbert-tone')

def get_sentiment_score(text):
    inputs = tokenizer(text, return_tensors='pt', truncation=True, padding=True, max_length=512)
    outputs = model(**inputs)
    logits = outputs.logits
    probs = torch.nn.functional.softmax(logits, dim=-1)
    
    positive_score = probs[0][0].item()
    # Scale the score to be between 0 and 1
    sentiment_score = positive_score
    
    return sentiment_score

def analyze_sentiment(row):
    text = row['Text']
    sentiment_score = get_sentiment_score(text)
    return sentiment_score

df['Sentiment Score'] = df.apply(analyze_sentiment, axis=1)
```

```python
bullish_patterns = ["double bottom", "triple bottom", "cup and handle", "rounding bottom", "falling wedge", "bull flag", "bullish pennant", "ascending triangle", "reverse head and shoulders", ]
bearish_patterns = ["double top", "triple top", "reverse cup and handle", "rising wedge", "bear flag", "descending triangle", "head and shoulders"]
```

```python
def check_patterns(text, patterns):
    for pattern in patterns:
        if pattern in text.lower():
            return True
    return False

df["Bullish"] = df["Text"].apply(lambda x: check_patterns(x, bullish_patterns))
df["Bearish"] = df["Text"].apply(lambda x: check_patterns(x, bearish_patterns))

df_textblob['Bullish'] = df_textblob['Lemmatized'].apply(lambda x: check_patterns(x, bullish_patterns))
df_textblob['Bearish'] = df_textblob['Lemmatized'].apply(lambda x: check_patterns(x, bearish_patterns))
```

```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_trader_colors(traders):
    unique_traders = list(set(traders))
    colors = plt.get_cmap('tab10', len(unique_traders))
    trader_color_map = {trader: colors(i) for i, trader in enumerate(unique_traders)}
    return trader_color_map

def create_btc_graph(topic_df):
    G = nx.DiGraph()

    topic_node = 'BTC'
    G.add_node(topic_node, label=topic_node, layer=0, color='gray', sentiment_score=0.5)
    
    topic_df = topic_df.sort_values(by='Timestamp')
    last_post_label = {}
    
    trader_colors = get_trader_colors(topic_df['Trader'])
    
    for _, row in topic_df.iterrows():
        post_label = f"Post_{row['Timestamp']}"
        title = row['Title']
        trader = row['Trader']
        post_type = row['Post type']
        sentiment_score = row['Sentiment Score']
        
        if sentiment_score > 0.5:
            node_color = 'green'
        elif sentiment_score < 0.5:
            node_color = 'red'
        else:
            node_color = 'gray'
        
        if post_type == 'Initial post':
            G.add_node(post_label, label=trader, layer=1, color=node_color, sentiment_score=sentiment_score)
            G.add_edge(post_label, topic_node, color='black')
            last_post_label[title] = post_label
        else:
            if title in last_post_label:
                previous_post_label = last_post_label[title]
                G.add_node(post_label, label=trader, layer=G.nodes[previous_post_label]['layer'] + 1, color=node_color, sentiment_score=sentiment_score)
                
                previous_color = G.nodes[previous_post_label]['color']
                edge_color = 'green' if previous_color == node_color else 'red'
                
                G.add_edge(post_label, previous_post_label, color=edge_color)
                last_post_label[title] = post_label
    
    # Writing to .bag file
    with open("BTC_graph_14.bag", "w") as file:
        for node, data in G.nodes(data=True):
            file.write(f"arg({node},{data['sentiment_score']}).\n")
        
        for u, v, data in G.edges(data=True):
            relation = 'sup' if G.nodes[u]['sentiment_score'] > 0.5 else 'att'
            file.write(f"{relation}({u},{v}).\n")
    
    pos = nx.multipartite_layout(G, subset_key="layer", align="vertical")
    
    plt.figure(figsize=(16, 12))
    
    node_colors = [G.nodes[node]['color'] for node in G.nodes]
    edge_colors = [G[u][v]['color'] for u, v in G.edges]
    
    nx.draw(G, pos, with_labels=True, labels=nx.get_node_attributes(G, 'label'), node_size=3000, node_color=node_colors, font_size=10, font_color="black", font_weight="bold", edge_color=edge_colors, arrows=True)
    plt.title('BTC Topic Graph')
    plt.show()

create_btc_graph(btc_df[:14])
```
